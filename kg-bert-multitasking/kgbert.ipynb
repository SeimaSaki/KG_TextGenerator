```
from bertviz import model_view
from transformers import BertTokenizer, BertModel
import torch
```

```
def call_html():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              "d3": "https://cdnjs.cloudflare.com/ajax/libs/d3/5.7.0/d3.min",
              jquery: '//ajax.googleapis.com/ajax/libs/jquery/2.0.0/jquery.min',
            },
          });
        </script>
        '''))
```

**Model view of link prediction multi-tasking model on a sample sentence pair**

```
model_version = 'bert-base-uncased'
do_lower_case = True
model_state_dict = torch.load('../pals/linkpred_2604/best_model.pth')
##model = BertModel.from_pretrained(model_version, state_dict=model_state_dict, output_attentions=True)
model = BertModel.from_pretrained(model_version, output_attentions=True)
tokenizer = BertTokenizer.from_pretrained(model_version, do_lower_case=do_lower_case)
sentence_a = "The cat sat on the mat"
sentence_b = "The cat lay on the rug"
inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='pt', add_special_tokens=True)
token_type_ids = inputs['token_type_ids']
input_ids = inputs['input_ids']
attention = model(input_ids, token_type_ids=token_type_ids)[-1]
##print('Attention: ', attention)
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list)
call_html()
model_view(attention, tokens)
```

**Model view of link prediction multi-tasking model on a question-answer pair from the QNLI dev dataset**

```
model_version = 'bert-base-uncased'
do_lower_case = True
model_state_dict = torch.load('../pals/linkpred_2604/best_model.pth')
##model = BertModel.from_pretrained(model_version, state_dict=model_state_dict, output_attentions=True)
model = BertModel.from_pretrained(model_version, output_attentions=True)
tokenizer = BertTokenizer.from_pretrained(model_version, do_lower_case=do_lower_case)
sentence_a = "What is the capital of Iran?"
sentence_b = "Tehran"
inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='pt', add_special_tokens=True)
token_type_ids = inputs['token_type_ids']
input_ids = inputs['input_ids']
attention = model(input_ids, token_type_ids=token_type_ids)[-1]
##print('Attention: ', attention)
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list)
call_html()
model_view(attention, tokens)
```

**Neuron view of link prediction multi-tasking model on a sample sentence pair**

```
from bertviz.transformers_neuron_view import BertModel, BertTokenizer
from bertviz.neuron_view import show
```

```
model_type = 'bert'
model_version = 'bert-base-uncased'
do_lower_case = True
model_state_dict = torch.load('../pals/linkpred_2604/best_model.pth')
##model = BertModel.from_pretrained(model_version)
model = BertModel.from_pretrained(model_version, state_dict=model_state_dict)
tokenizer = BertTokenizer.from_pretrained(model_version, do_lower_case=do_lower_case)
sentence_a = "The cat sat on the mat"
sentence_b = "The cat lay on the rug"
call_html()
show(model, model_type, tokenizer, sentence_a, sentence_b)
```

**Neuron view of link prediction multi-tasking model on a question-answer pair from QNLI dev dataset**

```
model_type = 'bert'
model_version = 'bert-base-uncased'
do_lower_case = True
model_state_dict = torch.load('../pals/linkpred_2604/best_model.pth')
##model = BertModel.from_pretrained(model_version)
model = BertModel.from_pretrained(model_version, state_dict=model_state_dict)
tokenizer = BertTokenizer.from_pretrained(model_version, do_lower_case=do_lower_case)
sentence_a = "San Jose is a city in the state of"
sentence_b = "California"
call_html()
show(model, model_type, tokenizer, sentence_a, sentence_b)
```
